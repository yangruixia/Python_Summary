```python
import os
import re
import json
import nltk
from nltk.tokenize import word_tokenize
import jieba.posseg as pseg

from collections import Counter
import numpy as np
```

```python
#读取文件内容，返回文件内容
def ReadContent(fname):
    with open(fname, 'r', encoding = 'utf-8')as file_object:
        content = file_object.read()
    file_object.close()
    return content

#将内容content写入名为file_name的文件
def WriteContent(file_name, content):
    with open(file_name, "w",encoding='utf-8') as f:
        json.dump(content, f,ensure_ascii=False,indent = 4)
    return ''
```

```python
#英文文本处理

#计算英文文章段落数目
def DESPC_en(content):
    j = 0
    temp = ''
    para_num = 1
    length = []
    for i in content:
        j += 1
        temp += i 
        if i == '\n':
            para_num += 1
            token = word_tokenize(temp)
            length.append(len(token))
            temp = ''
        elif j == len(content):
            token = word_tokenize(temp)
            length.append(len(token))
    return length,para_num

#计算英文文章段落的平均长度
def DESPL_en(content):
    text = DESPC_en(content)[0]
    para_num = DESPC_en(content)[1]
    my_len = 0
    for i in text:
        my_len += i
    average_len = my_len/para_num
    return average_len

#计算英文段落平均长度的标准差
def DESPLd_en(content):
    text = DESPC_en(content)[0]
    count_len = []
    for i in text:
        count_len.append(i)
    my_num = np.std(count_len, ddof = 1)    # 需要得到样本标准差
    return my_num

#上述信息进行整合并输出
def descraption_paragraph_en(fname):
    text_en = ReadContent(fname)
    my_len = DESPC_en(text_en)[1]
    paragraph_en_num = DESPL_en(text_en)
    paragraph_str = DESPLd_en(text_en)
    print('该英文文本的段落数为：', my_len)
    print('该英文文本段落的平均长度为：',paragraph_en_num)
    print('该英文文本段落平均长度的标准偏差为：',paragraph_str)
```

```python
#中文文本处理

#计算中文文章段落数目
def DESPC_zh(content):
    j = 0
    temp = ''
    para_num = 1
    length = []
    for i in content:
        j += 1
        if i == ' ':
            continue
        else:
            temp += i 
        if i == '\n':
            temp = temp.rstrip('\n')
            para_num += 1
            length.append(len(temp))
            temp = ''
        elif j == len(content):
            length.append(len(temp))
            temp = ''
    return length,para_num

#计算中文文章段落的平均长度
def DESPL_zh(content):
    text = DESPC_zh(content)[0]
    para_num = DESPC_zh(content)[1]
    my_len = 0
    for i in text:
        my_len += i
    average_len = my_len/para_num
    return average_len

#计算中文段落平均长度的标准差
def DESPLd_zh(content):
    text = DESPC_zh(content)[0]
    count_len = []
    for i in text:
        count_len.append(i)
    my_num = np.std(count_len, ddof = 1)    # 需要得到样本标准差
    return my_num

#上述信息进行整合并输出
def descraption_paragraph_zh(fname):
    text_zh = ReadContent(fname)
    my_len = DESPC_en(text_zh)[1]
    paragraph_zh_num = DESPL_zh(text_zh)
    paragraph_str = DESPLd_zh(text_zh)
    print('该中文文本的段落数为：', my_len)
    print('该中文文本段落的平均长度为：', paragraph_zh_num)
    print('该中文文本段落平均长度的标准偏差为：', paragraph_str)
```

```python
# 修改文本名称即可
descraption_paragraph_en('entest.txt')
descraption_paragraph_zh('zhtest.txt')
```
