```python
#IP地址归属地自动查询全代码
import requests
url = 'http://ip138.com/ips138.asp?ip='
try:
    r = requests.get(url+'192.168.255.2')
    r.raise_for_status()
    r.encoding = r.apparent_encoding
    print(r.text[-500:])
except:
    print('爬取失败')
```

```python
from pyquery import PyQuery as pq
```

```python
import requests
url = 'http://www.blcu.edu.cn'
try:
    r = requests.get(url)
    r.raise_for_status()
    r.encoding = r.apparent_encoding
    print(r.text)
except:
    print('爬取失败')
```

```python
type(r)
```

```python
html_text = r.text
```

```python
doc = pq(html_text)
```

```python
doc('head').text()
```

```python
doc('body').text()
```

```python
lis = doc('.zcontent li')#class
```

```python
for li in lis:
    print(pq(li).find('a').text())
```
