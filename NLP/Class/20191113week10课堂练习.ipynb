```python
from selenium import webdriver
from pyquery import PyQuery as pq
```

```python
dr = webdriver.Chrome()
```

```python
dr.get("http://www.baidu.com")
```

```python
dr.find_elements_by_id('kw')
```

```python
dr.find_element_by_id('kw').send_keys('杨瑞霞')
```

```python
dr.find_element_by_id('su').click()
```

```python
dr.find_element_by_id('content_left')
```

```python
s = dr.page_source
```

```python
print(s)
```

```python
doc = pq(s)
```

```python
lst = doc('#content_left .c-container h3 a em')
```

```python
print(lst)
```

```python
len(lst)
```

```python
lis = doc('#content_left .c-container h3 a').items()#.items会是一个生成器
print(type(lis))
for li in lis:
    print(li)
```

# 获取底部元素


```python
dr.find_element_by_id('page')
```

```python
f = dr.page_source
```

```python
doc2 = pq(f)
```

```python
lst2 = doc2('#page a')
```

```python
print(lst2)
```

```python
lis2 = doc('#page a').items()#.items会是一个生成器
print(type(lis2))
for li in lis2:
    print(li)
```
