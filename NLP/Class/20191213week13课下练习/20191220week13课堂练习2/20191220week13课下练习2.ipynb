```python
import pymongo
import pandas as pd

# 链接数据库
client = pymongo.MongoClient('localhost', 27017)
db = client["kekenet"]

# 查看数据库下所有聚集的名称
# coll_names = db.list_collection_names(session=None)

collection = db["体育"]

content = {}

for item in collection.find():    
    content = item
    break

zh = content['zh']
en = content['en']


# nltk进行词法和句法分析

import nltk


text = en.split(' ')
from nltk.util import ngrams
unigram = ngrams(text,1)
print('分词', [i for i in unigram])
bigram = ngrams(text,2)
print('两者分词', [i for i in bigram])
trigram = ngrams(text,3)
print('三者分词', [i for i in trigram])

from nltk.tokenize import sent_tokenize

text = en
sent_tokenize_list = sent_tokenize(text)
print(sent_tokenize_list)

from nltk.stem.porter import PorterStemmer
from nltk.stem.lancaster import LancasterStemmer
from nltk.stem import WordNetLemmatizer

porter_stemmer = PorterStemmer()
lancsater_stemmer = LancasterStemmer()
porter_stemmer.stem(en)
lancsater_stemmer.stem(en)

wordnet_lemmatizer = WordNetLemmatizer()
wordnet_lemmatizer.lemmatize(en)


# Spacy进行词法和句法分析

import spacy
nlp = spacy.load('en_core_web_sm')
text= en
# print(tex0t)
doc = nlp(text)
# 分句
list(doc.sents)
# 词性标注
for w in list(doc.sents)[0]:print(w,w.tag_)
all_tags = {w.pos: w.pos_ for w in doc}
print('词性标注', all_tags)
# 命名实体检测
print('命名实体检测', [ (w,w.label_)  for w in doc.ents])
# 名词性短语
print('名词性短语', [np for np in doc.noun_chunks])
```
